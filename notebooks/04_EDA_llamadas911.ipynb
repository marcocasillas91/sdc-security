```python
file = open("C:/Users/Valeria/OneDrive/datalab_sdc/sdc-security/datasets/catalogo-de-colonias.json")
nbhs_catalog = json.load(file)
```

```python
simple_feats = []
for colonia_feats in nbhs_catalog['features']:
    simple_feat = {}
    simple_feat.update(colonia_feats['properties'])
    simple_feat['coordinates'] = colonia_feats['geometry']['coordinates']
    simple_feats.append(simple_feat)
```

```python
parsed_catalog = pd.DataFrame(simple_feats)
parsed_catalog
```

Converting to upper case for convenience

```python
pip install unidecode
```

```python
from unidecode import unidecode


def rewrite(text):
    return unidecode(text).upper()

parsed_catalog['colonia'] = parsed_catalog['colonia'].apply(rewrite)
parsed_catalog['alc'] = parsed_catalog['alc'].apply(rewrite)
parsed_catalog.head()
```

```python
cuau_catalog = parsed_catalog[parsed_catalog.alc == 'CUAUHTEMOC'].copy()
cuau_catalog
```

```python
import re
cuau_calls['colonia_cierre'] = cuau_calls['colonia_cierre'].str.replace(r'\s+(I|II|III|IV|V|VI|VII|VIII)$', '', regex=True)
cuau_calls
```

```python
print('Calls dataframe before merge contains {} rows'.format(cuau_calls.shape[0]))
print(cuau_calls.isnull().sum())

print('Catalog dataframe before merge contains {} rows'.format(cuau_catalog.shape[0]))
print(cuau_catalog.isnull().sum())

merge_colonias_cuau = cuau_calls.merge(cuau_catalog, left_on = ['alcaldia_cierre', 'colonia_cierre'],
                                 right_on = ['alc','colonia'], how = 'left')

print('Merge dataframe contains {} rows'.format(merge_colonias_cuau.shape[0]))
print(merge_colonias_cuau.isnull().sum())
merge_colonias_cuau
```
