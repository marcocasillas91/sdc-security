### Closing code column
We won´t need duplicated calls

```python
print("Different 'codigo_cierre' column values: {}\n".format(calls.codigo_cierre.unique()))
values_to_drop = ['Duplicado']  #'Negativo'

# Create a mask of rows to drop
rows_to_drop = calls['codigo_cierre'].isin(values_to_drop)

print("Rows to drop:")
print(calls[rows_to_drop])

# Drop rows with any of the specified values in 'codigo_cierre'
calls = calls[~rows_to_drop]
print(calls.shape)
```

### Deleting call types that are not related to crime
For the propouses of this project we are interested only in certain types of emergency call: those related with crimes in public road.
We´ll base on columns: 'clas_con_f_alarma', 'categoria_incidente_c4' and 'incidente_c4', that go from the general to the particular describing the reason for the call, for example: DELITO - Agresion - A policia  

```python
#First most general category:'clas_con_f_alarma'

for item in calls.clas_con_f_alarma.unique():
    print(f"Count of {item} value: {(calls['clas_con_f_alarma'] == str(item)).sum()}")
```

```python
#We are not interested in false alarm
calls = calls[calls['clas_con_f_alarma'] != 'FALSA ALARMA']
print(calls.shape)
```

```python
#Second most general category: 'categoria_incidente_c4' 
for item in calls.categoria_incidente_c4.unique():
    print(f"Count of {item} value: {(calls['categoria_incidente_c4'] == str(item)).sum()}")
```

```python
#we are not interested in: Incendio, Daños por fenomeno natural o tercero, Derrame o fuga,  Contra la salud, Administrativas, Abandono,
# Explosion, Sismo, Accidente, Electorales

values_to_drop = ['Incendio', 'Danos por fenomeno natural o tercero', 'Derrame o fuga', 'Contra la salud', 
                  'Administrativas', 'Abandono', 'Explosion', 'Sismo', 'Accidente', 'Electorales']
rows_to_drop = calls['categoria_incidente_c4'].isin(values_to_drop)

print("Rows to drop:")
print(calls[rows_to_drop])

# Drop rows with any of the specified values in 'categoria_incidente_c4'
calls = calls[~rows_to_drop]
print(calls.shape)
```

```python
#The last most general category: 'categoria_incidente_c4' 
for item in calls.incidente_c4.unique():
    print(f"Count of {item} value: {(calls['incidente_c4'] == str(item)).sum()}")
```

```python
#we are interested in few categories
values_to_keep = ['Persona sospechosa', 'Otros', 'Persona extraviada', 'A policia', 
                  'Secuestro', 'Portacion de arma', 'Menor en riesgo', 'Por arma blanca', 
                  'Por golpes', 'Robo', 'Persona', 'Disparo', 'Agresion', 'Policia', 
                  'Sistema tecnologico de videovigilancia', 'Secuestro express', 'Danos', 'Vias del metro'
                  'Estrangulado']

# Keep rows with any of the specified values in 'categoria_incidente_c4'
calls = calls[calls['incidente_c4'].isin(values_to_keep)]
```

```python
print(calls.shape)
print("Different 'clas_con_f_alarma' column values: {}\n".format(calls.clas_con_f_alarma.unique()))
print("Different 'categoria_incidente_c4' column values: {}\n".format(calls.categoria_incidente_c4.unique()))
print("Different 'incidente_c4' column values: {}\n'".format(calls.incidente_c4.unique()))

print("Count of different 'clas_con_f_alarma' column values: {}\n".format(calls.clas_con_f_alarma.unique().size))
print("Count of different 'categoria_incidente_c4' column values: {}\n".format(calls.categoria_incidente_c4.unique().size))
print("Count of different 'incidente_c4' column values: {}\n".format(calls.incidente_c4.unique().size))
```

```python
# Get unique values in the 'incidente_c4' column for rows with specific 'categoria_incidente_c4'
for item in calls.categoria_incidente_c4.unique():
    calls_category_incident_c4 = calls[calls['categoria_incidente_c4'] == str(item)]
    print(f"Different incidents c4 values of {item} category: {calls_category_incident_c4['incidente_c4'].unique()}") 
```
